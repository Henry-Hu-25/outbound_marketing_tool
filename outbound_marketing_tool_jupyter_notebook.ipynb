## Setting up Contrastive Language-Image Pretraining and BM25

```python
from transformers import CLIPProcessor, CLIPModel
import torch
```

```python
model_name = "openai/clip-vit-base-patch32"
model = CLIPModel.from_pretrained(model_name)
processor = CLIPProcessor.from_pretrained(model_name)
```

```python
def clip_image_embedding(image_path):
    '''
    Get dense embeddings for images
    '''
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt")  # Preprocess the image and convert it to a tensor format suitable for the model
    with torch.no_grad():
        embedding=model.get_image_features(**inputs)
    # normalization for faster cosine similarity calculation
    return embedding/embedding.norm(dim=-1)
```

```python
def clip_text_embedding(query):
    '''
    Get dense embeddings for text
    '''
    # Preprocess the text query
    inputs = processor(
        text=[query], return_tensors="pt", padding=True, truncation=True
    )
    with torch.no_grad():
        embedding = model.get_text_features(**inputs)
    # normalization for faster cosine similarity calculation
    return embedding / embedding.norm(dim=-1)
```

```python
from pinecone_text.sparse import BM25Encoder
```

```python
bm25 = BM25Encoder()
bm25.fit(inventory_df['Description'])
```

```python
def bm25_encode_documents(description):
    '''
    Get sparse embeddings for text descriptions
    '''
    return bm25.encode_documents(description)
```

```python
def bm25_encode_queries(query):
    '''
    Get sparse embeddings for queries
    '''
    return bm25.encode_queries(query)
```

## Upserting Inventory Dataframe into Pinecone

```python
inventory_df.columns
```

```python
for _, row in inventory_df.iterrows():
    # Extracting Elements
    image_id = row["style"]
    image_path = row["Image"]
    image_description = row["Description"]

    # Image Embedding
    dense_embeds = clip_image_embedding(image_path).squeeze(0).tolist()

    # Text Embedding
    sparse_embeds = bm25_encode_documents(image_description)
    
    # Get metadata
    metadata = row.to_dict()
    if index.fetch([image_id])['vectors']:
        print(f"ID {image_id} already exists in the database. Skipping upsert.")
        continue
    else:
        index.upsert([{
            'id': image_id,
            'sparse_values': sparse_embeds,
            'values': dense_embeds,
            'metadata': metadata
        }])
```

```python
def hybrid_scale(sparse, dense, style_similarity=1):
    '''
    Hybrid vector scaling with alpha * dense + (1 - alpha) * sparse

    Args:
      - dense: array of floats
      - sparse: a dict of indices and values
      - alpha: 0 = sparse only, 1 = dense only
    '''
    if style_similarity < 0 or style_similarity > 1:
        raise ValueError("Alpha must be between 0 and 1")

    scaled_sparse = {
        "indices": sparse["indices"],
        "values": [
            value * (1 - style_similarity) for value in sparse["values"]
        ]
    }
    scaled_dense = [style_similarity * value for value in dense]
    return scaled_sparse, scaled_dense
```

## Hybrid Search Result Comparisons

```python
def collect_search_results(search_result):
    '''
    Collect search results for search visualizations

    Args:
      - search_result: a pinecone query response
    
    Return:
      - a list of style names
      - a list of image paths
      - a list of composition descriptions
    '''

    style_names = []
    image_paths = []
    composition_desc = []

    for match in search_result.matches:
        style_names.append(match["id"])
        image_paths.append(match["metadata"]["Image"])
        composition_desc.append(match['metadata']['Description'])
    
    return style_names, image_paths, composition_desc
```

```python
def show_matched_images(
        style_names, image_paths, composition_desc
):
    '''
    Display the matched styles

    Args:
      - style_names: a list of style names
      - image_paths: a list of image paths
      - composition_desc: a list of composition descriptions

    '''
    # Display images in a grid (3 per row)
    fig, axes = plt.subplots(
        nrows=(len(image_paths) + 2) // 3,  # Calculate rows based on images
        ncols=3, 
        figsize=(15, 12)  # Adjust size for better spacing
    )

    # Flatten axes for easy iteration
    axes = axes.flatten()

    # Plot each image
    for i, image_path in enumerate(image_paths):
        img = Image.open(image_path)
        axes[i].imshow(img)
        axes[i].axis("off")
        axes[i].set_title(style_names[i], fontsize=10)
        axes[i].text(
            0.5,
            -0.05, 
            "\n".join(textwrap.wrap(composition_desc[i], width = 50)),
            fontsize=8,
            ha="center",
            va="top",
            transform=axes[i].transAxes
        )

    plt.show()
```

```python
from langsmith import Client
from langchain.smith import RunEvalConfig, run_on_dataset
from langsmith.evaluation import LangChainStringEvaluator, evaluate
from langsmith.schemas import DataType
```

```python
client = Client()
dataset_name = "outbound_cold_emails"
```

```python
if not client.has_dataset(dataset_name = dataset_name):
        dataset = client.create_dataset(
            dataset_name=dataset_name,
            data_type=DataType.kv,
            description="This dataset contains outbound email drafts"
        )
```

```python
client.create_example(
    dataset_name = dataset_name,
    inputs = {"inputs": outbound_email.content}
)
```

```python
evaluator_conciseness = LangChainStringEvaluator(
        "criteria",
        config={
            "criteria": "conciseness"
        }
)
evaluator_clarity = LangChainStringEvaluator(
        "criteria",
        config={
            "criteria": {
                "clarity": "Does the email clearly communicate the value proposition?"
            }
        }
)
evaluator_personalization = LangChainStringEvaluator(
        "criteria",
        config={
            "criteria": {
                "personalization": "Is the email properly personalized to the client?"
            }
        }
)
evaluator_call_to_action = LangChainStringEvaluator(
        "criteria",
        config={
            "criteria": {
                "call_to_action": "Is there a clear and compelling call to action?"
            }
        }
)
# client.run_on_dataset(
#     dataset_name=dataset_name,
#     llm_or_chain_factory=llm_groq,
#     project_name="test-cold-experiment-43",
#     verbose=True
# )
```
